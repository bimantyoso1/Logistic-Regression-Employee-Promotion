Data Pre-Processing (sederhana)

```python
#Copy Dataset
df_cp = df.copy()
```

```python
#Urutkan index
df_cp = df_cp.reset_index(drop = True)
df_cp
```

```python
df_cp.info()
```

```python
#Kumpulkan semua kolom kategorikal
#karena algoritma hanya dapat menerima numerikal
columns_cat = df_cp.select_dtypes(include = ['object']).columns.to_list()
columns_cat
```

kita gunakan encoder untuk merubah kategorikal menjadi numerikal

```python
label_encoders = {} # dictionary untuk menyimpan kumpulan encoder dari kolom2 categorical

for col in columns_cat:
  label_encoder = LabelEncoder() # buat objek label encoder untuk setiap kolom categorical
  label_encoder.fit(df_cp[col]) # label encoder melakukan training untuk isi kolom kategorical
  df_cp[col] = label_encoder.transform(df_cp[col]) # label encoder melakukan transformasi, timpa isi kolom tsb sebelumnya
  label_encoders[col] = label_encoder # simpan objek encoder di dictionary
  pickle.dump(label_encoder, open('label_encoder_' + col + '.pkl', 'wb')) # simpan dalam file pickle di server gcolab
```

```python
df_cp
```

```python
#Drop kolom yang tidak sesuai
X = df_cp.drop(columns = ['employee_id', 'is_promoted'])
y = df_cp['is_promoted'] #kolom target
```

```python
#Split data set untuk training 80% dan testing 20% dan melakukan pemisahan data random
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 32)
```

**ML Modelling**

```python
logreg = LogisticRegression() # buat objek logistic regression
```

```python
logreg
```

```python
#membuat model Log Reg
model_logreg = logreg.fit(X_train, y_train)
```

```python
y_pred_train = model_logreg.predict(X_train)
y_pred_test = model_logreg.predict(X_test)
```

```python
training_acc = accuracy_score(y_train, y_pred_train)
test_acc = accuracy_score(y_test, y_pred_test)
```

Check Score Akurasi

```python
print('training accuracy:', training_acc)
print('testing accuracy:', test_acc)
```
